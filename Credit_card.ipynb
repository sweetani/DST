```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Credit = pd.read_csv('default.csv')
```

```python
Credit.info()
```

```python
Credit.describe()
```

```python
Credit.head()
```

```python
Credit['outstanding_sep']=Credit['Sep_BILL_AMT1']-Credit['Sep_PAY_AMT1']
Credit['outstanding_Aug']=Credit['Aug_BILL_AMT2']-Credit['Aug_PAY_AMT2']
Credit['outstanding_July']=Credit['July_BILL_AMT3']-Credit['July_PAY_AMT3']
Credit['outstanding_June']=Credit['June_BILL_AMT4']-Credit['June_PAY_AMT4']
Credit['outstanding_May']=Credit['May_BILL_AMT5']-Credit['May_PAY_AMT5']
Credit['outstanding_April']=Credit['April_BILL_AMT6']-Credit['April_PAY_AMT6']
```

```python
Credit['Outstanding']=Credit['outstanding_sep']+Credit['outstanding_Aug']+Credit['outstanding_July']+Credit['outstanding_June']+Credit['outstanding_May']+Credit['outstanding_April']
```

```python
#Overall outstanding amount trends
sns.distplot(Credit['Outstanding'])
```

```python
#Number of customers with outstanding amount (in different outstanding amount buckets)
n = (Credit['Outstanding'].max()-Credit['Outstanding'].min()) // 1000
bucket = pd.cut(Credit['Outstanding'],n)
bucket
```

```python
#Age of outstanding amount analysis
Credit['Age_out']=Credit['Sep_PAY_0']+Credit['Aug_PAY_2']+Credit['July_PAY_3']+Credit['June_PAY_4']+Credit['May_PAY_5']+Credit['April_PAY_6']
```

```python
sns.jointplot(x='Age_out',y='Outstanding',data=Credit)
```

```python
#Is there any relationship between in outstanding amount / trend with respect to age, geo, education, marriage, credit limit
Relation=Credit[['Outstanding','AGE','EDUCATION','MARRIAGE']]
Relation.corr()
sns.heatmap(Relation.corr(),annot=True)
```

```python
#Does outstanding amount / trend affect the default behavior in next month.
sns.jointplot(x='default payment next month',y='Outstanding',data=Credit,kind='reg')
```
